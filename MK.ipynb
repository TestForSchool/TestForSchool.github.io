```python
#Loads Spreadsheets
import pandas as pd
import matplotlib.pyplot as plt
# Load data from CSV files into separate DataFrames
body_stats_df = pd.read_csv('MarioKartBodyStats.csv')
driver_stats_df = pd.read_csv('MarioKartDriverStats.csv')
glider_stats_df = pd.read_csv('MarioKartGliderStats.csv')
wheel_stats_df = pd.read_csv('MarioKartWheelStats.csv')
```

```python
# Adds Speed+Acceleration For Vehicles 
#print(body_stats_df)
body_stats_df['total_speed_acceleration'] = body_stats_df['Ground_Speed'] + \
 body_stats_df['Acceleration']
sorted_body_stats_df = body_stats_df.sort_values(
    by='total_speed_acceleration', ascending=False)
#print(sorted_body_stats_df[['Vehicle', 'total_speed_acceleration']])

body_highest_speed_acceleration = sorted_body_stats_df[sorted_body_stats_df['total_speed_acceleration']== sorted_body_stats_df['total_speed_acceleration'].max()]
highest_miniturbo_body = body_highest_speed_acceleration[body_highest_speed_acceleration['Mini-Turbo']== body_highest_speed_acceleration['Mini-Turbo'].max()]
print(highest_miniturbo_body[['Vehicle','Ground_Speed','Acceleration', 'total_speed_acceleration', 'Mini-Turbo']])
```

```python
# Adds Speed+Acceleration For Drivers
driver_stats_df['total_speed_acceleration'] = driver_stats_df['Ground_Speed'] + \
    driver_stats_df['Acceleration']
sorted_driver_stats_df = driver_stats_df.sort_values(
    by='total_speed_acceleration', ascending=False)
#print(sorted_body_stats_df[['Vehicle', 'total_speed_acceleration']])
# Calculate a new column 'total_speed_acceleration' by adding 'speed' and 'acceleration'

# Find the vehicles with the highest speed + acceleration
driver_highest_speed_acceleration = sorted_driver_stats_df[sorted_driver_stats_df['total_speed_acceleration']
                                                  == sorted_driver_stats_df['total_speed_acceleration'].max()]

# Now, from the subset of vehicles with the highest speed + acceleration, find the ones with the highest miniturbo stat
highest_miniturbo_driver = driver_highest_speed_acceleration[driver_highest_speed_acceleration['Mini-Turbo']
                                               == driver_highest_speed_acceleration['Mini-Turbo'].max()]

# Print the vehicles with the highest miniturbo among the highest speed + acceleration
#print(highest_miniturbo[['Driver', 'total_speed_acceleration', 'Mini-Turbo']])
```

```python
# Adds Speed+Acceleration For Gliders
glider_stats_df['total_speed_acceleration'] = glider_stats_df['Ground_Speed'] + \
    glider_stats_df['Acceleration']
sorted_glider_stats_df = glider_stats_df.sort_values(
    by='total_speed_acceleration', ascending=False)
#print(sorted_body_stats_df[['Vehicle', 'total_speed_acceleration']])
# Calculate a new column 'total_speed_acceleration' by adding 'speed' and 'acceleration'

# Find the Gliders with the highest speed + acceleration
highest_speed_acceleration = sorted_glider_stats_df[sorted_glider_stats_df['total_speed_acceleration']
                                                           == sorted_glider_stats_df['total_speed_acceleration'].max()]

# Now, from the subset of vehicles with the highest speed + acceleration, find the ones with the highest miniturbo stat
highest_miniturbo = highest_speed_acceleration[highest_speed_acceleration['Mini-Turbo']
                                                      == highest_speed_acceleration['Mini-Turbo'].max()]

# Print the vehicles with the highest miniturbo among the highest speed + acceleration
#print(highest_miniturbo[['Glider', 'total_speed_acceleration', 'Mini-Turbo']])
```

```python
# Adds Speed+Acceleration For Gliders
wheel_stats_df['total_speed_acceleration'] = wheel_stats_df['Ground_Speed'] + \
    wheel_stats_df['Acceleration']
sorted_wheel_stats_df = wheel_stats_df.sort_values(
    by='total_speed_acceleration', ascending=False)
#print(sorted_body_stats_df[['Vehicle', 'total_speed_acceleration']])
# Calculate a new column 'total_speed_acceleration' by adding 'speed' and 'acceleration'

# Find the Gliders with the highest speed + acceleration
highest_speed_acceleration = sorted_wheel_stats_df[sorted_wheel_stats_df['total_speed_acceleration']
                                                    == sorted_wheel_stats_df['total_speed_acceleration'].max()]

# Now, from the subset of vehicles with the highest speed + acceleration, find the ones with the highest miniturbo stat
highest_miniturbo_tire = highest_speed_acceleration[highest_speed_acceleration['Mini-Turbo']
                                               == highest_speed_acceleration['Mini-Turbo'].max()]

# Print the vehicles with the highest miniturbo among the highest speed + acceleration
#print(highest_miniturbo[['Tire', 'total_speed_acceleration', 'Mini-Turbo']])
```

```python
import matplotlib.pyplot as plt

# Extract data for the graphs
# Speed and acceleration for the best combinations
best_combinations = [sorted_driver_stats_df, sorted_body_stats_df, sorted_glider_stats_df, sorted_wheel_stats_df
                     ]
combinations_names = ["Mr. Scooty/Biddybuggy", "Rollers/AzulRollers", "Waluigi/Roy", "Cloudglider/Paperglider"]
speed = [combo['total_speed_acceleration'].values[0]
         for combo in best_combinations]
acceleration = [combo['total_speed_acceleration'].values[1]
                for combo in best_combinations]
miniturbo = [combo['Mini-Turbo'].values[0] for combo in best_combinations]

# Create a bar graph for Speed and Acceleration
plt.figure(figsize=(10, 5))
plt.bar(combinations_names, speed, label='Speed+ Acceleration')
plt.xlabel('Parts')
plt.ylabel('Speed and Acceleration')
plt.title('Best Combinations - Speed and Acceleration')
plt.legend()

# Show the first graph
plt.show()

# Create a bar graph for Mini-Turbo
plt.figure(figsize=(10, 5))
plt.bar(combinations_names, miniturbo, color='purple')
plt.xlabel('Parts')
plt.ylabel('Mini-Turbo')
plt.title('Best Combinations - Mini-Turbo')

# Show the second graph
plt.show()
```
